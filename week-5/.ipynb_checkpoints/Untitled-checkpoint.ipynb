```python
import tkinter as tk
from tkinter import messagebox

math_entry = None
eng_entry = None
phy_entry = None
comp_entry = None
civic_entry = None

mass_math_entry = None
mass_eng_entry = None
mass_phy_entry = None
mass_arts_entry = None
mass_literature_entry = None

admitted = []
not_admitted = []

root = tk.Tk()
root.title("Verification Window")
root.geometry("500x300")

name_label = tk.Label(root, text="Name:")
name_label.pack()
name_entry = tk.Entry(root)
name_entry.pack()

jamb_label = tk.Label(root, text="JAMB score:")
jamb_label.pack()
jamb_entry = tk.Entry(root)
jamb_entry.pack()

interview_label = tk.Label(root, text="Did you pass the interview? (y/n)")
interview_label.pack()
interview_entry = tk.Entry(root)
interview_entry.pack()

def button_click_cos():
    global math_entry, eng_entry, phy_entry, comp_entry, civic_entry
    name = name_entry.get()
    jamb = int(jamb_entry.get())
    interview = interview_entry.get().upper()

    if jamb >= 250 and interview == "Y":
        window = tk.Toplevel()
        window.geometry("500x300")
        window.title("Computer Science Registration")

        math_label = tk.Label(window, text="Mathematics Grade:")
        math_label.pack()
        math_entry = tk.Entry(window)
        math_entry.pack()

        eng_label = tk.Label(window, text="English Grade:")
        eng_label.pack()
        eng_entry = tk.Entry(window)
        eng_entry.pack()

        phy_label = tk.Label(window, text="Physics Grade:")
        phy_label.pack()
        phy_entry = tk.Entry(window)
        phy_entry.pack()

        comp_label = tk.Label(window, text="Computer Science Grade:")
        comp_label.pack()
        comp_entry = tk.Entry(window)
        comp_entry.pack()

        civic_label = tk.Label(window, text="Civic Education Grade:")
        civic_label.pack()
        civic_entry = tk.Entry(window)
        civic_entry.pack()

        cos_conf_button = tk.Button(window, text="Admit", command=button_comp_conf)
        cos_conf_button.pack()
    else:
        not_admitted.append((name, jamb, interview))
        messagebox.showerror("Error", "You're not eligible")

def button_click_mass():
    global mass_math_entry, mass_eng_entry, mass_phy_entry, mass_arts_entry, mass_literature_entry
    name = name_entry.get()
    jamb = int(jamb_entry.get())
    interview = interview_entry.get().upper()

    if jamb >= 230 and interview == "Y":
        window = tk.Toplevel()
        window.geometry("500x300")
        window.title("Mass Communication Registration")

        math_label = tk.Label(window, text="Mathematics Grade:")
        math_label.pack()
        mass_math_entry = tk.Entry(window)
        mass_math_entry.pack()

        eng_label = tk.Label(window, text="English Grade:")
        eng_label.pack()
        mass_eng_entry = tk.Entry(window)
        mass_eng_entry.pack()

        phy_label = tk.Label(window, text="Physics Grade:")
        phy_label.pack()
        mass_phy_entry = tk.Entry(window)
        mass_phy_entry.pack()

        arts_label = tk.Label(window, text="Arts Grade:")
        arts_label.pack()
        mass_arts_entry = tk.Entry(window)
        mass_arts_entry.pack()

        literature_label = tk.Label(window, text="Literature Grade:")
        literature_label.pack()
        mass_literature_entry = tk.Entry(window)
        mass_literature_entry.pack()

        mass_conf_button = tk.Button(window, text="Admit", command=button_mass_conf)
        mass_conf_button.pack()
    else:
        not_admitted.append((name, jamb, interview))
        messagebox.showerror("Error", "You're not eligible for Mass Communication.")

def button_comp_conf():
    global math_entry, eng_entry, phy_entry, comp_entry, civic_entry
    math = math_entry.get().upper()
    eng = eng_entry.get().upper()
    phy = phy_entry.get().upper()
    comp = comp_entry.get().upper()
    civic = civic_entry.get().upper()

    if math in ("A", "B", "C") and eng in ("A", "B", "C") and phy in ("A", "B", "C") and comp in ("A", "B", "C") and civic in ("A", "B", "C"):
        admitted.append((name_entry.get(), int(jamb_entry.get()), interview_entry.get().upper(), "Computer Science"))
        messagebox.showinfo("Info", "You are Admitted to Computer Science.")
    else:
        messagebox.showerror("Error", "You cannot be admitted to Computer Science.")

def button_mass_conf():
    global mass_math_entry, mass_eng_entry, mass_phy_entry, mass_arts_entry, mass_literature_entry
    math = mass_math_entry.get().upper()
    eng = mass_eng_entry.get().upper()
    phy = mass_phy_entry.get().upper()
    arts = mass_arts_entry.get().upper()
    literature = mass_literature_entry.get().upper()

    if math in ("A", "B", "C") and eng in ("A", "B", "C") and phy in ("A", "B", "C") and arts in ("A", "B", "C") and literature in ("A", "B", "C"):
        admitted.append((name_entry.get(), int(jamb_entry.get()), interview_entry.get().upper(), "Mass Communication"))
        messagebox.showinfo("Info", "You are Admitted to Mass Communication.")
    else:
        not_admitted.append((name_entry.get(), int(jamb_entry.get()), interview_entry.get().upper(), "Mass Communication"))
        messagebox.showerror("Error", "You cannot be admitted to Mass Communication.")


cos_button = tk.Button(root, text="Computer Science", command=button_click_cos)
cos_button.pack()

mass_button = tk.Button(root, text="Mass Communication", command=button_click_mass)
mass_button.pack()

root.mainloop()

print("Admitted candidates: {}".format(admitted))
print("Candidates that weren't admitted: {}".format(not_admitted))
```
